## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
# This cell is for CODE (numbers, a query, a check).
import os, sys, subprocess

IN_COLAB = "google.colab" in sys.modules
REPO_URL = "https://github.com/flyrank-bih/flyrank-ml-internship-starter"
REPO_DIR = "flyrank-ml-internship-starter"

if IN_COLAB:
    if not os.path.isdir(REPO_DIR):
        subprocess.run(["git", "clone", "--depth", "1", REPO_URL, REPO_DIR], check=True)
    os.chdir(REPO_DIR)
    subprocess.run([sys.executable, "-m", "pip", "install", "-q", "-r", "requirements.txt"], check=True)
else:
    while not os.path.isdir("data/raw") and os.getcwd() != "/":
        os.chdir("..")

import pandas as pd, numpy as np

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

# Same filters the starter pipeline applies: visible, mature content, deduplicated.
df = df[(df["impressions_90d"] > 0) & (df["content_age_days"] >= 90)].drop_duplicates("content_id")
n = len(df)

declining_with_demand = ((df["trend_direction"].str.lower() == "down") & (df["impressions_90d"] >= 100)).sum()
low_ctr_visible = ((df["impressions_90d"] >= 500) & (df["avg_position"] > 0)
                    & (df["avg_position"] <= 20) & (df["ctr"] < 0.5)).sum()
stale_visible = ((df["days_since_last_update"] >= 180) & (df["impressions_90d"] >= 500)).sum()

print(f"Rows after filters (visible, mature, deduplicated): {n:,}")
print(f"declining_with_demand (down-trending, impressions_90d >= 100): {declining_with_demand:,}  "
      f"({100*declining_with_demand/n:.1f}%)")
print(f"low_ctr_visible_page (high impressions, top-20 position, CTR < 0.5%): {low_ctr_visible:,}  "
      f"({100*low_ctr_visible/n:.1f}%)")
print(f"stale_visible_page (not updated in 6+ months, still getting real traffic): {stale_visible:,}  "
      f"({100*stale_visible/n:.1f}%)")

```
